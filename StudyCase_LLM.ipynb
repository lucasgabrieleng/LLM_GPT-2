```python
# Load the tokenizer associated with the model
dsa_tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

## Loading Pre-trained Model

```python
# Load Model
modelo_dsa = GPT2LMHeadModel.from_pretrained("gpt2-large", pad_token_id = dsa_tokenizer.eos_token_id)
```

```python
modelo_dsa
```

## Generating Text with the LLM

```python
# Initial Text
prompt = 'What is Artificial Intelligence'
```

```python
# Tokenize the prompt text
input_ids = dsa_tokenizer.encode(prompt, return_tensors = 'pt')
```

```python
# Visualization
input_ids
```

```python
# Generating text from initial prompt
texto_gerado = modelo_dsa.generate(input_ids,
                                   max_length = 100,
                                   num_beams = 5,
                                   no_repeat_ngram_size = 2,
                                   early_stopping = True)
```

```python
texto_gerado
```

## Decoding the result with Tokenizer

```python
print(dsa_tokenizer.decode(texto_gerado[0], skip_special_tokens = True))
```
